We load in Example.csv file from the working directory defined by the variable wd. Next, using some attributes of the data frame called df to do basic data analysis, like head and describe.

```python
df = pd.read_csv(wd+'/Example.csv') 
print df.head()
print df.describe()

print '\n\n'
print df.isnull().values.any() 
print 'So, no missing value exists.' # check the missing values. FALSE: no missing value. 
```

Using plt.scatter to do a scatter plot of the first column vs the second column in df. 

```python
plt.scatter(df.iloc[:100,1],df.iloc[:100,2]) # a scatter plot of x vs y.
plt.xlabel('one')
plt.ylabel('two')
plt.show()
```

Using plt.hsit to do a histgram of the third column in df. Also you may change the setting of arugment bins in the function to increase the resolution. 

```python
plt.hist(df.iloc[:,3])
plt.xlabel('three')
plt.ylabel('frequency')
plt.show()
```

```python
plt.hist(df.iloc[:,3], bins=50, rwidth=0.5, alpha = 0.5, lw=3,color= 'k')
plt.xlabel('three')
plt.ylabel('frequency')
plt.show()
```

Finally, we add a pie chart of two pre-specified variables: F and M in the last column of the data frame called df. Note that we should compute the frequencies of F and M before we plot the pie. 

```python
labels_inkey = df['key'].unique()

print labels_inkey

labels_inkey_ = {x:0 for x in labels_inkey}

# count the frequency of each item. 

for item in df['key']:
    labels_inkey_[item] += 1

# obtain the frequencies of F and M. 
sizes = [labels_inkey_[item] for item in ['F','M']]

explode = [0]*len(sizes)
explode[0] = 0.1  # explode 1st slice

plt.pie(sizes, explode=explode, labels=['F','M'], autopct='%1.1f%%', 
        shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```
